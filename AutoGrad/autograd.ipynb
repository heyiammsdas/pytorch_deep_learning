## Why AutoGrad exists? 

>> AutoGrad = automatic gradient(Partial differention) engine 

In neural network, we need gradients to update weights 

W_new = W_old - (learning_rate) * (dL/dW) 

Gradient: rate of change of loss w.r.t weight (dL/dW) 

>> Manually computing this derivative is complex , error prone, tedious 
So PyTorch introduces AutoGrad

>> It automatically computes gradient using - 
chain rule 

dL/dW =  dL/dy * dy/dW
