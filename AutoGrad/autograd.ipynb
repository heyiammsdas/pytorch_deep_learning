## Why AutoGrad exists? 

>> AutoGrad = automatic gradient(Partial differention) engine 

In neural network, we need gradients to update weights 

W_new = W_old - (learning_rate) * (dL/dW) 

Gradient: rate of change of loss w.r.t weight (dL/dW) 

>> Manually computing this derivative is complex , error prone, tedious 
So PyTorch introduces AutoGrad

>> It automatically computes gradient using - 
chain rule 

dL/dW =  dL/dy * dy/dW 

computational graph 

x ----\
       (*) ----\
w ----/         (+) ----> y
b -------------/









Now , let x = 2 
and y = x**3 

now i wanna differentiate y wrt x 




```python
import torch 

x = torch.tensor(2.0, requires_grad=True);

```

```python
y = x**3
y
```

```python
y.backward() # i have differentiate wrt x
```

```python
x.grad
```

### Partial Differentiation 

x->y->z 

x=3
y = x**3
z = sin(y) 

then what is the dz/dx ? 


```python
x = torch.tensor(3.0 , requires_grad=True) 

y = x**3 

z = torch.sin(y) 

z.backward() 

print(x.grad) 
```

### One Important Thing 

PyTorch only stores gradient of leaf tensors 

x(leaf)->y(intermidiate)->z(root) 

Error y.grad z.grad

PyTorch internally computes using chain rule dz/dx = dz/dy * dy/dx

```python
# detach() => It cuts the tensor from computation graph 

y = x * 2
z = y.detach() * 3

# z.backward()

# Gradient won't go back to x 

```

### Zeroing Gradient 

in pytorch gradient accumulates (adds up) by default 

so we have to reset the gradient to zero 


```python
import torch

x = torch.tensor(3.0 , requires_grad=True) 

y = x**3 

z = torch.sin(y) 

z.backward() 

x.grad

print(x.grad)

x.grad = None

print(x.grad)
```
