# what is PyTorch ? and Why do we need this ? 

Pytorch is a tool(library) used to build and train machine learning and deep learning models

without PyTorch 

- we have to do everything manually 
- matrix operations and slow and complex
- finding gradients is hard

with pytorch : 

- pytorch do the math automatically 
- handles gradients 
- works fast (uses gpu , cpu) 



## What are the tensors? 

Tensors are n dimentional array 

### 0-D tensor : actually scaler 
eg. : 0 
1 
4

### 1-D tensor : vectors 

eg. [2,4,6] 

used let say one row of [iq , cgpa , state] 

it is 1-D tensor but the dimension of the vector is 3 (becoz 3 elements are there)


### 2-D tensor : matrix 
eg. [
    [2,4,5],
    [7,4,3]
]

shape : (2,3) == (row , column)

used : 
if there is 1000 rows in the dataset of students of iq | cgpa | state | placed
for inputs there is 1000 vectors of [iq , cgpa , state]

### 3-D tensor : (3rd order tensor)
 [[
    [2,3,5] ,
    [5,4,1]
] , [
    [7,2,3],
    [9,8,7]
]]

example : time series data of stock 

higest | lowest 

365 days and 10 years 

(10 , 365 , 2) 

### 4-D tensors 

R G B layers , 50 images

(50,3,255,255)

### 5-D tensors 

videos in 30fps for 60 second

(60 , 30 , 3 , 255, 255)

### Operations

```python
ran1 = torch.rand(3,4) 
ran2 = torch.rand(3,4) 

print(ran1) 
print(ran2)
```

```python
# addition 
print(ran1+ran2) 

# substraction
#division
#multiplication

```

```python
# matrix multiplication 

x = torch.rand(3,4)*10 
y = torch.rand(4,4) *10 

c = torch.matmul(x,y) 

print(c)
```

```python
# indexing 

x = torch.Tensor([1,2,3,4,5,6])

print(x[4])

## 2D indexing 

x = torch.Tensor([[2,3,4], [4,5,6]]) 



print(x[0][2]) # from [2,3,4] take 2nd element  /(0th,1st,2nd)/ 
```

```python
# dot product 

x = torch.Tensor([1,2,4,5]) 
y = torch.Tensor([4,5,6,7])

c = torch.dot(x,y) 
print(c) 
```

```python
#determinant 

u = torch.rand(3,3) * 10 

print(torch.det(u))
```

```python
#Inplace Operation 


## NOT IN-PLACE
x = torch.Tensor([1,2,3]) 

y = x + 1 

print(x) 
print(y)

## IN-PLACE


torch.sqrt_(x)

print(x) 


# functions ending with underscore (_)  are in place mean less memory use 
# disadvange in autograd  because pytorch in original values for backprop

# add_() , mul_() 

```

```python
## Reshape 

a = torch.randint(size=(4,4) , low=1 , high=10) 

b = a.reshape(4,2,2) 

print(b) 

## flatten 

c = a.flatten()
print(c)


## Unsqueeze use to add new dimension


x = torch.Tensor([1,2,3])

y = x.unsqueeze(0) 
print(y)

z = x.unsqueeze(1)

print(z)

## Permute 
# used to rearrage dimension 

n = torch.randn(2,3,4) 
y = n.permute(1,0,2) 

# shape change from (2,3,4) -> (3,2,4)


## Numpy to Tensor 

import numpy as np 

arr = np.array([1,2,3]) 

x = torch.from_numpy(arr) 

print(x) 

## tensor to numpy 

arr = x.numpy() 

print(arr)

```
